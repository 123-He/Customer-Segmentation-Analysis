```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体：黑体（Windows 自带）
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel("./Online Retail.xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['Totalsum'] = df['Quantity']*df['UnitPrice']
```

```python
curday = df['InvoiceDate'].max()+ dt.timedelta(days=1)
rfm = df.groupby(by='CustomerID').agg(
    {
        'InvoiceDate':lambda x:(curday-x.max()).days,
        'CustomerID':'count',
        'Totalsum':'sum'
    }
)
rfm
```

```python
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'CustomerID': 'Frequency',
    'Totalsum': 'Monetary'
}, inplace=True)
```

```python
rfm.head()
```

```python
r_labels = range(4,0,-1)
rfm['r'] = pd.qcut(rfm['Recency'],q=4,labels=r_labels)
f_labels = range(1,5)
rfm['f'] = pd.qcut(rfm['Frequency'].rank(method='first'),q=4,labels=f_labels)
m_labels = range(1,5)
rfm['m'] = pd.qcut(rfm['Monetary'],q=4,labels=m_labels)
rfm['rfm_score'] = rfm[['r','f','m']].sum(axis=1)
rfm
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(rfm['rfm_score'], kde=True, bins=10)
plt.title('Customer RFM Score Distribution')
plt.show()
```

```python
def assign_segment(row):
    if row['rfm_score'] >= 10:
        return 'Champions (超级会员)'
    elif row['rfm_score'] >= 8:
        return 'Loyal (忠诚客户)'
    elif row['rfm_score'] >= 6:
        return 'Potential (潜力股)'
    else:
        return 'At Risk (需挽留)'
rfm['Segment'] = rfm.apply(assign_segment,axis=1)
```

```python
rfm
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(
    data=rfm,
    x='Recency', 
    y='Frequency',
    hue='Segment',
    palette='viridis',
    s=100,
    alpha=0.6
)
plt.title('RFM Customer Segmentation (Recency vs Frequency)')
plt.xlabel('Recency (Days since last purchase)')
plt.ylabel('Frequency (Number of purchases)')
plt.legend(title='Customer Segment')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
sclar = StandardScaler()
rfm_sclar = sclar.fit_transform(rfm[['Recency','Frequency','Monetary']])
kmeans = KMeans(n_clusters=4,random_state=42)
kmeans.fit(rfm_sclar)
rfm['KMeans_Cluster'] = kmeans.labels_
rfm
```

```python
rfm['KMeans_Cluster'].value_counts()
```

```python
rfm.reset_index(inplace=True)
```

```python
print("\n--- K-Means 自动聚类的结果统计 ---")
print(rfm.groupby('KMeans_Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'CustomerID': 'count'
}))
```

```python
# 5. 画图对比 (用 AI 的分类结果上色)
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=rfm, 
    x='Recency', 
    y='Frequency', 
    hue='KMeans_Cluster', # 注意这里改成了 KMeans_Cluster
    palette='tab10', 
    s=100, 
    alpha=0.6
)
plt.title('K-Means Clustering Results')
plt.show()
```

```python
rfm.columns
```

```python
rfm_log = rfm[['Recency', 'Frequency', 'Monetary']].copy()
```

```python
rfm_log['Recency'] = np.log1p(rfm_log['Recency'])
rfm_log['Frequency'] = np.log1p(rfm_log['Frequency'])
rfm_log['Monetary'] = np.log1p(rfm_log[rfm_log['Monetary']>0]['Monetary'])
```

```python
rfm_log.dropna(inplace=True)
```

```python
scalar = StandardScaler()
rfm_log_scalar = scalar.fit_transform(rfm_log)
kmeans = KMeans(n_clusters=4,random_state=42)
kmeans.fit(rfm_log_scalar)
rfm.loc[rfm_log.index,'KMeans_Cluster_Log'] = kmeans.labels_
rfm
```

```python
rfm.dropna(inplace=True)
rfm
```

```python
cluster_summary = rfm.groupby('KMeans_Cluster_Log').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'Monetary':['mean','count'],
}).round(1)
cluster_summary.index = ['一般用户','超级vip用户','vip用户','需要尝试挽回的用户']
cluster_summary
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(
    data=rfm,
    x='Recency',
    y='Frequency',
    hue='KMeans_Cluster_Log',
    s=60,
    alpha=0.7
)
plt.title('K-Means Segmentation (Log Scale Visualization)')
plt.xlabel('Recency (Days) - Log Scale')
plt.ylabel('Frequency (Times) - Log Scale')
plt.show()
```

# 基于 K-Means 聚类结果的营销建议

## 1. 超级VIP用户 (R=12, M=6493)
- **特征**：极度活跃，高消费。
- **策略**：不需要打折！应该提供“尊享服务”（如专属客服、新品优先购买权）。给折扣反而浪费利润。

---

## 2. 一般用户 (R=94, M=1324)
- **特征**：消费能力不错（比VIP还高！），但最近3个月没来了，有流失风险。
- **策略**：重点召回对象。推送“老客回归福利”或“新品推荐”，提醒他们回来看看。

---

## 3. VIP用户 (R=18, M=597)
- **特征**：非常活跃（最近18天刚来），但单价低（M仅597）。可能是喜欢买便宜小东西的用户。
- **策略**：提升客单价 (Upselling)。推荐“满减凑单”或“组合套餐”，引导他们多花点钱。

---

## 4. 需要挽回的用户 (R=183, M=284)
- **特征**：半年没来，消费低，属于边缘用户。
- **策略**：低成本群发。发送大额优惠券（如满100减50），能拉回一个是一个，拉不回也不亏。
